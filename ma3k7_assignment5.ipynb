```python
# import relevant libraries
import matplotlib.pyplot as plt
import numpy as np

import matplotlib
import matplotlib as mpl
```

```python
# supporting functions to move the position
def up(x,y):
    return x-1, y

def left(x,y):
    return x, y-1

def right(x,y):
    return x, y+1

def down (x,y):
    return x+1, y
```

```python
def next_spiral(spiral):
    #initialise new matrix extended by 1 row and 1 column on either sides to make room for new spiral
    old_size = np.shape(spiral)[0]
    new_size = np.shape(spiral)[0]+2
    new_spiral = np.zeros((new_size, new_size))
    
    new_spiral[1:-1, 1:-1] = spiral
    pos = np.shape(spiral)
    
    # moves made to generate new spiral values
    new_spiral[right(pos[0], pos[1])] = new_spiral[pos]+1
    pos = right(pos[0], pos[1])
    
    for _ in range(old_size):
        new_spiral[up(pos[0], pos[1])] = new_spiral[pos]+1
        pos = up(pos[0], pos[1])
        
    for _ in range(old_size+1):
        new_spiral[left(pos[0], pos[1])] = new_spiral[pos]+1
        pos = left(pos[0], pos[1])
        
    for _ in range(old_size+1):
        new_spiral[down(pos[0], pos[1])] = new_spiral[pos]+1
        pos = down(pos[0], pos[1])
    
    for _ in range(old_size+1):
        new_spiral[right(pos[0], pos[1])] = new_spiral[pos]+1
        pos = right(pos[0], pos[1])
    
    return new_spiral
    
    
```

```python
# test the function to extend the example
a = np.array([[5, 4, 3],
             [6, 1, 2],
             [7, 8, 9]])
```

```python
next_spiral(a)
```

```python
# use the function to build a spiral from scratch
def get_spiral(N):
    iters = int(np.floor(N/2))
    matrix = np.array([[1]])
    
    for _ in range(iters):
        matrix = next_spiral(matrix)
        
    return matrix
```

```python
# use the function to generate the example
a_spiral = get_spiral(7)
```

```python
b_spiral = get_spiral(11)
```

```python
# showing the spiral in a plot
fig, ax = plt.subplots()
im = ax.imshow(a_spiral)

for i in range(np.shape(a_spiral)[0]):
    for j in range(np.shape(a_spiral)[0]):
        text = ax.text(j, i, a_spiral[i, j],
                       ha="center", va="center", color="w")
        
ax.set_title("Spiral of Size 7")
plt.show()
```

```python
# showing another bigger spiral in a plot
fig, ax = plt.subplots()
im = ax.imshow(b_spiral)


for i in range(np.shape(b_spiral)[0]):
    for j in range(np.shape(b_spiral)[0]):
        text = ax.text(j, i, b_spiral[i, j],
                       fontsize= 8, ha="center", va="center", color="w")
        
ax.set_title("Spiral of Size 11")
plt.show()
```
